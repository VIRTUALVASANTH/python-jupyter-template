### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(10) * 10
y = np.random.rand(10) * 10

try:
    if len(x) != len(y):
        raise ValueError("x and y arrays must have the same length.")

    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, color='red', alpha=0.7)

    # Annotate points
    for i in range(len(x)):
        plt.text(x[i], y[i], f"({x[i]:.1f}, {y[i]:.1f})")

    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.title("Scatter Plot Example")
    plt.grid(True, linestyle='--', alpha=0.7)

    plt.show()

except Exception as e:
    print(f"Error: {e}")

```
