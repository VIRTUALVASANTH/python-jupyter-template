### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axes[0, 0].plot(x, y, color='b', label=r'$y = x^2$')
axes[0, 0].set_title("Line Plot")
axes[0, 0].grid(True)
axes[0, 0].legend()

# Bar Plot
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Plot")

# Scatter Plot
axes[1, 0].scatter(x_scatter, y_scatter, color='red', marker='o')
axes[1, 0].set_title("Scatter Plot")
for i in range(len(x_scatter)):
    axes[1, 0].text(x_scatter[i], y_scatter[i], f"({x_scatter[i]}, {y_scatter[i]})", fontsize=10, ha='right')

# Pie Chart
axes[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
